# Основы программирования в Python

## Домашнее задание 7

1. Извлечь с помощью пакетов `requests` и `BeautifulSoup` данные с html-страницы. В каком формате сохранять данные на предварительном этапе (список списков, кортежей, словарь и т.п.) - совершенно не важно.
2. Сохранить извлеченные данные в формате датафрейма `pandas`. Присвоить столбцам таблицы внятные названия. Обязательно преобразовать типы данных в столбцах (чтобы числовые данные имели формат `float` или `int`). Если в таблице присутствует «мусор», отфильтровать ненужные строки или удалить их с помощью метода `.drop()`.
3. Экспортировать полученные данные в Excel-файл.

Какие данные? Данные по Международному индексу счастья и его компонент за 2012 год. [Ссылка](https://ru.wikipedia.org/wiki/%D0%9C%D0%B5%D0%B6%D0%B4%D1%83%D0%BD%D0%B0%D1%80%D0%BE%D0%B4%D0%BD%D1%8B%D0%B9_%D0%B8%D0%BD%D0%B4%D0%B5%D0%BA%D1%81_%D1%81%D1%87%D0%B0%D1%81%D1%82%D1%8C%D1%8F) на страницу (нужная таблица - справа). Каждая строка в итоговом датафрейме должна соответствовать одной стране. В таблице должно быть 5 столбцов: *название страны*, *HPI*, *Удовлетворенность жизнью*, *Ожидаемая продолжительность жизни, Экологический след*.
