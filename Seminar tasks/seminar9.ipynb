## Семинар 16

В этом семинаре вам предстоит потренироваться в использовании API ВКонтакте для извлечения информации из этой социальной сети. 
Основная задача следующая: собрать данные о друзьях одного или нескольких ваших друзей, то есть получить таблицу, в которой будут представлены 
- id друга вашего друга (далее - пользователя)
- имя пользователя
- фамилия пользователя
- пол пользователя
- город пользователя
- время последнего визита пользователя

Для вашего удобства задание разбито на мини-шаги.

**Шаг 1.** Импортируйте пакет `vk_requests` под псевдонимом `vk`. Если пакет не установлен - предварительно установите его.

**Шаг 2.** Создайте переменную `api`, с помощью которой вы сможете получить доступ к базе данных ВК (посредством вашего токена доступа).

**Шаг 3.** Найдите в [документации](https://vk.com/dev/methods) метод, который позволяет на основе id пользователя получить список его друзей с некоторой дополнительной информацией о них. Этот шаг не требует написания кода - просто запомните, какой метод вам нужен.

**Шаг 4.** Выберите среди своих друзей одного человека, друзей которого вам было бы интересно изучить. Определите id его страницы ВК. Поместите этот id в переменную `user` с типом `int`.

**Шаг 5.** Определите, какой из параметров выбранного на шаге 3 метода отвечает за получение расширенной информации о друзьях пользователя. Этот шаг не требует написания кода - просто запомните, какой параметр вам нужен.

**Шаг 6.** Используя выбранный на шаге 3 метод и параметр с шага 5 достаньте список друзей пользователя, id которого лежит в переменной `user`. Необходимо, чтобы кроме самих id друзей выдавалась также информация об их поле, городе и времени последнего визита ВК. Сохраните результат в переменную `user_friends`. Пока ничего не нужно предобрабатывать, пусть это будет список из многих словарей. Кроме того, сохраните общее количество друзей в переменную `user_friends_count`.

**Шаг 7.** Рассчитайте, какой процент друзей пользователя вам удалось охватить на этом этапе. Выведите результат на экран.

**Шаг 8.** Создайте более понятную структуру данных с информацией о друзьях пользователя - список списков. В каждом вложенном списке должна идти следующая информация о друге пользователя: id, имя, фамилия, пол, название города и время последнего визита ВК (в формате UNIX). Поместите результат в переменную `user_friends_clean`. У некоторых пользователей может не быть каких-то данных - возникнет ошибка. Предусмотрите эту ситуацию, использовав конструкцию try-except. Пусть в итоговую базу попадут только те пользователи, у которых есть вся нужная нам информация.

**Шаг 9.** Определите, информацию о скольких пользователях вам таким образом удалось собрать.

**Шаг 10.** Оформите полученный результат в датафрейм. Назовите его `data`, дайте адекватные названия колонок.

**Шаг 11.** Проведите небольшую предобработку: создайте колонку `DateTime` с человеческим отображением времени последнего визита. Затем разделите эту колонку на две: в одной будет дата визита (`Date`), в другой - время (`Time`). Удалите две исходные колонки (самую первую с UNIX-временем и колонку `DateTime`).

**Шаг 12.** Проведите небольшую предобработку: перекодируйте данные в колонке с полом из 1-2 в женский-мужской. Заментие старую колонку новой.

**Шаг 13.** Проведите небольшую предобработку: создайте переменную, которая будет отражать, в Москве человек находится или нет, на основе значений из колонки с городом. В этой колонке должны быть значения "Москва" и "Не Москва", сама колонка пусть называется `Moscow`.

**Шаг 14.** Экспортируйте получившийся датафрейм в excel-файл.

**Шаг 15.** Расслабьтесь и отдохните - Вы хорошо потрудились.
