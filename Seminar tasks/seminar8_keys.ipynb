**Шаг 8.** Структурируйте этот код, превратив его в прекрасный суп, и выведите на экран. Кстати, чтобы код отображался ещё лучше, можно при выводе использовать метод `.prettify()`, т.е. `print(soup.prettify())`.

**Шаг 9.** С помощью инструментов BeautifulSoup-a найдите в коде страницы однотипные теги, в каждом из которых содержится полная информация об одной из программ. Сохраните список из этих тегов в переменную `programs`.

```python
programs = soup.find_all('li', {'class': 'ais-InfiniteHits-item'})
programs
```

**Шаг 10.** Возьмите первый тег из полученного списка `programs`, сохраните его в переменную `course`. Выведите на экран с помощью `print(course.prettify())` - в этом теге должна содержаться информация о степени под названием Master of Business Administration.

```python
course = programs[0]
print(course.prettify())
```

**Шаг 11.** Извлеките из тега в переменной `course` название программы, поместите в переменную `title`.

```python
title = course.find('h2').get_text()
title
```

**Шаг 12.** Извлеките из тега в переменной `course` ссылку на программу, достройте её до абсолютной ссылки (прибавьте к адресу основной адрес сайта) и поместите в переменную `link`.

```python
link = 'https://www.coursera.org' + course.find('a').get('href')
link
```

**Шаг 13.** Извлеките из тега в переменной `course` организатора программы, поместите в переменную `organizer`.

```python
organizer = course.find('span', {'class': 'partner-name'}).get_text()
organizer
```

**Шаг 14.** Извлеките из тега в переменной `course` тип программы (Degree / Professional Certificate), поместите в переменную `program_type`.

```python
program_type = course.find('div', {'class':'product-type-row horizontal-box'}).get_text()
program_type
```

**Шаг 15.** Наконец, извлеките из тега в переменной `course` краткое описание программы, поместите в переменную `description`.

```python
description = course.find('p', {'class':'tagline'}).get_text()
description
```

**Шаг 16.** Поместите переменные `title`, `link`, `organizer`, `program_type`, `description` в список `info_per_course`.

```python
info_per_course = [title, link, organizer,
                  program_type, description]
```

**Шаг 17.** Осуществите шаги 11-16 для всех программ на страничке с выдачей поиска, пройдясь циклом `for` по всем тегам из переменной `programs` и собирая информацию по ним в список списков `info_per_page`. Лайфхак - если назовёте переменную в цикле как `course` (`for course in programs`), то менять не придётся практически ничего ;) Ошибки обработайте с помощью try-except.

```python
info_per_page = []
for course in programs:
    title = course.find('h2').get_text()
    link = 'https://www.coursera.org' + course.find('a').get('href')
    organizer = course.find('span', {'class': 'partner-name'}).get_text()
    program_type = course.find('div', {'class':'product-type-row horizontal-box'}).get_text()
    try:
        description = course.find('p', {'class':'tagline'}).get_text()
    except:
        description = None
    info_per_course = [title, link, organizer,
                  program_type, description]
    info_per_page.append(info_per_course)
info_per_page
```

**Шаг 18.** Оформите полученный список списков в датафрейм с понятными названиями колонок.

```python
import pandas as pd
data = pd.DataFrame(info_per_page, columns = ['Title', 'URL', 'Organizer',
                                              'Type', 'Description'])
data
```

**Шаг 19.** Экспортируйте получившийся датафрейм в excel-файл.

```python
data.to_excel('Degress_Coursera.xlsx')
```
